```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(data)
print(df)
```

```python
#with pandas
print(data["calories"])
plt.plot(df)
plt.show
```

```python
#witout pandas
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.plot(data["calories"], label= "Calories")
plt.plot(data["duration"], label = "Duration")
plt.legend()
plt.show()
```

```python
#sales
data = pd.read_csv('Sales.csv')
```

```python
#head and tail and loc
print(data.tail())
print(data.tail(3))

print(data.head())
print(data.head(3))

print(data.loc[9])
```

```python
#prep data
id_sales = data.groupby("Date")['total'].sum()
id_sales.plot(kind='area')
plt.legend()
plt.ylabel("Sales Amount")
plt.show()
```
